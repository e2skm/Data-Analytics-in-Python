### Determine the proportion of male and female employees in each Stream.

```python
# Create a pivot table for proportions
gender_proportion_by_stream = capaciti_EOH_ADSA_df.pivot_table(index='Stream',columns='Gender',aggfunc='size',fill_value=0)
```

```python
gender_proportion_by_stream['Total'] = gender_proportion_by_stream.sum(axis=1)
```

```python
gender_proportion_by_stream['Male'] = round(gender_proportion_by_stream['Male'] / gender_proportion_by_stream['Total'] * 100,2)
```

```python
gender_proportion_by_stream['Female'] = round(gender_proportion_by_stream['Female'] / gender_proportion_by_stream['Total'] * 100,2)
```

```python
gender_proportion_by_stream[['Male','Female']]
```

# Find Outliers based on IQ scores

```python
# Find Q1
q1 = capaciti_EOH_ADSA_df['IQ'].quantile(0.25)
print('First Quatile:',q1)
```

```python
# Find Q3
q3 = capaciti_EOH_ADSA_df['IQ'].quantile(0.75)
print('Third Quatile:',q3)
```

```python
# Get Interquatile Range
IQR = q3 - q1
print('Interquatile Range:',IQR)
```

```python
# Get lower bound
lower_bound = q1 - 1.5 * IQR
print('Lower Bound:',lower_bound)
```

```python
# Get upper bound
upper_bound = q3 + 1.5 * IQR
print('Upper Bound:',upper_bound)
```

```python
# Use upper and lower bounds to get Outliers
outliers = capaciti_EOH_ADSA_df[(capaciti_EOH_ADSA_df['IQ'] >= upper_bound) | (capaciti_EOH_ADSA_df['IQ']<= lower_bound)]
outliers
```

```python
# Find top 10 candidates with the highest IQ
sorted_IQ = capaciti_EOH_ADSA_df['IQ'].sort_values(ascending=False)
highest_candidate_IQ = sorted_IQ.head(10)
print(highest_candidate_IQ)
```

```python
# Find buttom 10 candidates with the lowest IQ
lowest_candidate_IQ = sorted_IQ.tail(10)
lowest_candidate_IQ.reset_index(drop=True,inplace=True)
print(lowest_candidate_IQ)
```
