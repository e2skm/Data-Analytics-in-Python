### Import all the necessary Modules

```python
import pandas as pd
```

```python
import os
```

```python
import numpy as np
```

```python

np.random.seed(0)
```

### Generate data

```python
num_of_candidates = 50
```

```python
candidate_ids = [ 'CitiCan' + str(candidate).zfill(2) for candidate in range(1,num_of_candidates)]
```

```python
qualifications = np.random.choice(['Information Technology','Computer Science','Information Sytems','Computer Systems Engineering','Matric','Other'],size=num_of_candidates)
```

```python
level = np.random.randint(4,8,size=num_of_candidates)
```

```python
age = np.random.randint(21,36,size=num_of_candidates)
```

```python
monthlyStipend = 4950.00
```

```python
stream = np.random.choice(['Data Analytics','Cloud','Full Stack','Cyber Security',' '],size=num_of_candidates)
```

```python
genderint = np.random.randint(1,3,size=num_of_candidates)
```

```python
intelligence = np.random.randint(80,197,size=num_of_candidates)
```

### Store Generated data in a DataFrame

```python

capaciti_EOH_ADSA_data = pd.DataFrame({
  'CandidateID': np.random.choice(candidate_ids, size=num_of_candidates),
  'Qualifications': qualifications,
  'NQF Level': level,
  'Age': age,
  'Stipend': monthlyStipend,
  'Stream': stream,
  'Gender': genderint,
  'IQ': intelligence
})
```
