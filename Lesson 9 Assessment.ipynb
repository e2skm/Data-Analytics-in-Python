```python
# 3.Determine the proportion of male and female employees in each department.
gender_propotion_by_department = employee_performance_df.pivot_table(index='Department',columns='Gender',aggfunc='size',fill_value=0)
```

```python
gender_propotion_by_department['Total'] = gender_propotion_by_department.sum(axis=1)
```

```python
gender_propotion_by_department['Male_Propotion'] =  np.round((gender_propotion_by_department['Male'] / gender_propotion_by_department['Total']*100),2)
```

```python
gender_propotion_by_department['Female_Propotion'] = np.round((gender_propotion_by_department['Female'] / gender_propotion_by_department['Total'] *100),2)
```

```python
gender_propotion_by_department[['Male_Propotion', 'Female_Propotion']]
```

```python
# 4.Calculate the correlation between years of experience and monthly salary. Interpret the correlation coefficient.
experience_and_salary_correlation = employee_performance_df['YearsExperience'].corr(employee_performance_df['MonthlySalary'])
print('The Correlation between Employee Years Experience and Monthly Salary:',experience_and_salary_correlation)
```

```python
# 5.Identify outliers in the monthly salary using appropriate statistical methods.
q1 = employee_performance_df['MonthlySalary'].quantile(0.25)
q3 = employee_performance_df['MonthlySalary'].quantile(0.75) 
iqr = q3 - q1
```

```python
print('First Quatile: ',q1,'\nThird Quatile: ',q3,'\nInterquatile Range: ',iqr)
```

```python
lower_bound = q1 - 1.5 * iqr
print('Lower Bound:',lower_bound)
```

```python
upper_bound = q3 + 1.5 * iqr
print('Upper Bound:', upper_bound)
```

```python
outliers = employee_performance_df[(employee_performance_df['MonthlySalary'] < lower_bound) | (employee_performance_df['MonthlySalary'] > upper_bound)] 
outliers
```

```python
# 6.Compare the average performance ratings across different departments using descriptive statistics.
performance_rating_by_department = employee_performance_df.groupby('Department')['PerformanceRating'].mean()
performance_rating_by_department.round(2).sort_values(ascending=False)
```

```python
# Find Highest Salary by Gender
highest_salary_by_gender = employee_performance_df.groupby('Gender')['MonthlySalary'].max()
highest_salary_by_gender.sort_values(ascending=False)
```
