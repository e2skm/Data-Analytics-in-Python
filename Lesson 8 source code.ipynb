```python
import pandas as pd
```

```python
import os
```

```python
import numpy as np
```

```python
import random
```

```python
from datetime import datetime, timedelta
```

```python
np.random.seed(0)
```

```python
customer_ids = ['Cust' + str(i).zfill(3) for i in range(1,501)]
```

```python
products = ['Product A','Product B', 'Product C', 'Product D']
```

```python
start_date = datetime(2020,1,1)
```

```python
end_date = datetime(2020,12,31)
```

```python
dates = [start_date + timedelta(days=random.randint(0,(end_date - start_date).days))
         for _ in range(500)]
```

```python
purchase_amounts = np.random.randint(10,200, size=500)
```

```python
products_purchased = [np.random.choice(products) for _ in range(500)]
```

```python
purchase_data = pd.DataFrame({
  'CustomerID': np.random.choice(customer_ids,size=500),
  'Dates': dates,
  'Products': products_purchased,
  'PurchaseAmounts': purchase_amounts
})
```

```python
purchase_data.to_csv(r'C:\Users\Itumeleng.Sekoma\Downloads\PurchaseData.csv', index=False)
```
