## Import all the necessary liabraries

```python
import pandas as pd
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
df_sales = pd.read_csv(r'C:\Users\Itumeleng.Sekoma\Desktop\Test777777\.ipynb_checkpoints\sales_data.csv')
```

# Now calculate total sales

```python
total_sales = df_sales['Sales'].sum()
print("Total Sales for the Year 2023:", total_sales)
```

# Calculate average sales per day

```python
average_sales_per_day = df_sales.groupby('Date')['Sales'].sum().mean().round(2)
print("Average Sales per Day:", average_sales_per_day)
```

# Identify the top-performing product


```python
top_performing_product = df_sales.groupby('Product')['Sales'].sum().idxmax()
print("Top-Performing Product:", top_performing_product)
```

# Calculate sales by region

```python
sales_by_region = df_sales.groupby('Region')['Sales'].sum().sort_values(ascending=False)
```

```python
colors = ['g','b','orange','r']
```


# Create bar chart

```python
plt.figure(figsize=(10, 6))
bars = sales_by_region.plot(kind='bar', color=colors)
for bar in bars.patches:
    plt.text(bar.get_x() + bar.get_width() / 2, 
             bar.get_height() + 1000,               
             f'{bar.get_height()}',              
             ha='center',                        
             fontsize=10,                        
             fontfamily='arial',                 
             fontweight='bold')                      
plt.title('Sales by Region in Descending Order', fontfamily='arial', fontweight='bold')
plt.xlabel('Region', fontfamily='arial')
plt.ylabel('Total Sales', fontfamily='arial')
plt.show()
```

# Determine the day with the highest sales

```python
day_with_highest_sales = df_sales.loc[df_sales['Sales'].idxmax()]['Date']
print("Day with the Highest Sales:", day_with_highest_sales)
```

# Calculate the percentage of total sales by region

```python
total_sales_by_region = df_sales.groupby('Region')['Sales'].sum().sort_values(ascending=False)
percentage_sales_by_region = round((total_sales_by_region / total_sales) * 100,2)
print("Percentage of Total Sales by Region:")
print(percentage_sales_by_region)
```

# Add a 'Month' column to the dataframe

```python
df_sales['Month'] = pd.DatetimeIndex(df_sales['Date']).month
```


# Identify the month with the highest total sales

```python
month_with_highest_sales = df_sales.groupby('Month')['Sales'].sum().idxmax()
print("Month with the Highest Total Sales:", month_with_highest_sales)
```

### Calculate Other Metrics

```python
top_product_sales = df_sales.groupby('Product')['Sales'].sum().max()
max_sales_day = df_sales.groupby('Date')['Sales'].sum().idxmax()
max_sales_value = df_sales.groupby('Date')['Sales'].sum().max()
month_with_highest_sales = df_sales.groupby('Month')['Sales'].sum().idxmax()
month_with_highest_sales_value = df_sales.groupby('Month')['Sales'].sum().max()
```
