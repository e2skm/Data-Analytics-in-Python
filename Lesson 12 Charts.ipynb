```python
import numpy as np
import matplotlib.pyplot as plt
# Data
months = np.arange(1, 13)
revenue = np.array([1000, 1200, 1300, 1100, 900, 1000, 1500, 1600, 1700, 1800, 1900, 
2000])
average_revenue = np.mean(revenue)
# Plot
fig, ax1 = plt.subplots()
ax1.barh(months, revenue, color='b')
ax1.set_xlabel('Month')
ax1.set_ylabel('Revenue', color='b')
ax1.tick_params('y', colors='b')
ax2 = ax1.twinx()
ax2.plot(months, [average_revenue]*12, color='r')
ax2.set_ylabel('Average Revenue', color='r')
ax2.tick_params('y', colors='r')
plt.title('Monthly Revenue vs. Yearly Average')
plt.show()
```

```python
# Sample data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Product_A': [10, 15, 20, 25, 30, 35],
    'Product_B': [15, 20, 25, 30, 35, 40],
    'Product_C': [20, 25, 30, 35, 40, 45]
}

# Creating a DataFrame
import pandas as pd
df = pd.DataFrame(data)

# Plotting multiple lines
plt.figure(figsize=(10, 6))

plt.plot(df['Month'], df['Product_A'], label='Product A', marker='o' ,color='gold')
plt.plot(df['Month'], df['Product_B'], label='Product B', marker='s', color='skyblue')
plt.plot(df['Month'], df['Product_C'], label='Product C', marker='d', color='grey')

plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import plotly.graph_objects as go

# Sample data
data = {'categories': ['Sales', 'Returns', 'Marketing', 'R&D', 'Profit'],
        'values': [500, -50, -100, -80, 270]}

# Creating the waterfall chart
fig = go.Figure(go.Waterfall(
    x = data['categories'],
    y = data['values'],
    connector = {"line":{"color":"rgb(63, 63, 63)"}},
    increasing = {"marker":{"color":"green"}},
    decreasing = {"marker":{"color":"red"}},
    totals = {"marker":{"color":"blue"}}
))

fig.update_layout(
    title = "Waterfall Chart",
    xaxis_title = "Categories",
    yaxis_title = "Value"
)

fig.show()
```
